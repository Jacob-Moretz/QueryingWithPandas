Columns such as 'Deaths' are not being recognized as numeric due to the type of the data being Objects, and not ints.

```python
removed = df[(df['Deaths'] != 'Suppressed') & (df['Crude_Rate'] != 'Unreliable')]
df = removed
print(removed.head(20))
```

```python
df['Deaths'] = df['Deaths'].astype(float)
df['Crude_Rate'] = df['Crude_Rate'].astype(float)
pd.options.mode.copy_on_write = True 
```

## Task 6

```python
df_pop = df.groupby('State').agg({'Population': 'max'}).sort_values('Population', ascending= False)
df_pop.head(5)
```

 ## Task 7

```python
state_with_highest_pop = df_pop.index[0]
df_state = df[df['State'] == state_with_highest_pop]
top_5_year = df_state.nlargest(5, 'Deaths')[['Year' , 'Deaths']]
top_5_year = top_5_year.groupby('Year').agg({'Deaths' : 'max'})
top_5_year
```
